```python
import os
import numpy as np
import matplotlib.pyplot as plt
sr = 44.1*10e2
```

```python
rootDir = '../touchDesigner/data/generated/'
allIRs = os.listdir(rootDir)
allIRs
```

```python
for irFile in allIRs:
    fullPath = os.path.join(rootDir,irFile)
    ir = np.load(fullPath)
    N = len(ir)
    T = N/sr
    n = np.arange(N)
    t = n/sr
    plt.plot(t,ir)
    plt.xlim([0,0.1])
    yrang = 0.025
    plt.ylim([-yrang,yrang])
    
    plt.xlabel("Time [sec]")
    plt.title(irFile)
    plt.grid()
    plt.show()
```
