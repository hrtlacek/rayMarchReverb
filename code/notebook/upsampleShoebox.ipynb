```python
from scipy.io import wavfile
```

```python
import scipy.signal as sig
```

```python
sr,orig = wavfile.read('../touchDesigner/data/scenes/shoebox2/RIRs/wav/rir1.wav')
```

```python
ir = orig[:,0]
plot(ir)
xlim([0,1000])
```

```python
N = int((len(ir)/sr)*44100)
```

```python
rir4410 = sig.resample(ir,N)
```

```python
plot(rir4410)
xlim([0,(1000/sr)*44100])
title("resampled")
show()
```
